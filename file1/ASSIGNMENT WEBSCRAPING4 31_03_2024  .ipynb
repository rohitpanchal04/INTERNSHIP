# 01 QUESTION

```python
#OPENING CHROME
driver = webdriver.Chrome()
```

```python
#OPENING URL
driver.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')
```

```python
Most_Viewed_Videos=driver.find_element(By.XPATH,'//table["sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"][2]')
```

```python
Name=[]
Artist=[]
Views=[]
Upload_date=[]
```

```python
name=driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[1]')
for i in name:
    namme=i.text
    Name.append(namme)

artist=driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[2]')
for i in artist:
    art=i.text
    Artist.append(art)

view=driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[3]')
for i in view:
    views=i.text
    Views.append(views)

Date=driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[4]')
for i in Date:
    date=i.text
    Upload_date.append(date)
```

```python
print(len(Name),len(Artist),len(Upload_date),len(Views))
```

```python
#MAKING DATAFRAME OF SCRAPING DATA
Most_Viewed_VideosDF=pd.DataFrame({"Name":Name,"Artist":Artist,"Views":Views,"Upload_date":Upload_date})
```

```python
Rank =[]
State =[]
GSDP_21_22 =[]
GSDP_22_23 =[]
Share_21_22 =[]
GDP =[]
```

```python
#OPENING CHROME
driver=webdriver.Chrome()
```

```python
#OPENING URL
driver.get("https://archive.ics.uci.edu/")
```

```python
#CLICKING ON DATASET
dataset=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/footer/div[3]/a[2]')
dataset.click()
```

```python
Dataset_Name=[]
Dataset_Types=[]
Task=[]
Attribute_types=[]
No_of_Instance=[]
No_of_Attribute=[]
Year=[]
```

```python
Name=driver.find_elements(By.XPATH,'//h2[@class="truncate text-primary"]')
for i in Name:
    name=i.text
    Dataset_Name.append(name)

Types=driver.find_elements(By.XPATH,'//p[@class="truncate"]')
for i in Types:
    type=i.text
    Dataset_Types.append(type)

task=driver.find_elements(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[2]/div[2]/div/div/div/div/div[1]')
for i in task:
    task=i.text
    Task.append(task)

Attribute=driver.find_elements(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[2]/div[2]/div/div/div[2]/div/div[2]/span')
for i in Attribute:
    attr=i.text
    Attribute_types.append(attr)

Instances=driver.find_elements(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[2]/div[2]/div/div[1]/div[2]/div/div[3]/span')
for i in Instances:
    instance=i.text
    No_of_Instance.append(instance)

No_Attribute=driver.find_elements(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[2]/div[2]/div/div[1]/div[2]/div/div[4]/span')
for i in No_Attribute:
    No_attr=i.text
    No_of_Attribute.append(No_attr)

date=driver.find_elements(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[2]/div[2]/div/div/div/table/tbody/tr/td[3]')
for i in date:
    date=i.text
    Year.append(date)
```

```python
#MAKING DATAFRAME
Machine_LearningDF=pd.DataFrame({"Dataset_Name":Dataset_Name,"Dataset_Types":Dataset_Types,"Task":Task,"Attribute_types":Attribute_types,
                                 "No_of_Instance":No_of_Instance,"No_of_Instance":No_of_Instance,"Date":Year})
Machine_LearningDF
```
