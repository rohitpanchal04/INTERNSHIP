TOP 5 BATSMEN OF TEST (WEBSCRAPING BY SELENIUM)

```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
import time
```

```python
driver=webdriver.Chrome()
```

```python
driver.get('https://www.espncricinfo.com/records/most-runs-in-career-223646')
```

```python
Player_name=[]
Span=[]
matches=[]
Innings=[]
NotOut=[]
Runs=[]
High_Score=[]
Average=[]
Strike_Rate=[]
Hundreds=[]
Fifties=[]
Total_Fours=[]
Total_Sixes=[]
```

```python
player=driver.find_elements(By.XPATH,'//a[@class="ds-inline-flex ds-items-start ds-leading-none"]')
for i in player:
    name=i.text
    Player_name.append(name)
    Player_name=Player_name[0:107]
    
span=driver.find_elements(By.XPATH,'//*[@id="main-container"]/div[4]/div[1]/div[3]/div[1]/div[1]/div/div[1]/div[2]/table/tbody/tr/td[2]')
for i in span:
    span0=i.text
    Span.append(span0)
    
matche=driver.find_elements(By.XPATH,'//*[@id="main-container"]/div[4]/div[1]/div[3]/div[1]/div[1]/div/div[1]/div[2]/table/tbody/tr/td[3]')
for i in matche:
    match=i.text
    matches.append(match)
    
inning=driver.find_elements(By.XPATH,'//*[@id="main-container"]/div[4]/div[1]/div[3]/div[1]/div[1]/div/div[1]/div[2]/table/tbody/tr/td[4]')
for i in inning:
    ing=i.text
    Innings.append(ing)

Notout=driver.find_elements(By.XPATH,'//*[@id="main-container"]/div[4]/div[1]/div[3]/div[1]/div[1]/div/div[1]/div[2]/table/tbody/tr/td[5]')
for i in Notout:
    nt=i.text
    NotOut.append(nt)
    
run=driver.find_elements(By.XPATH,'//*[@id="main-container"]/div[4]/div[1]/div[3]/div[1]/div[1]/div/div[1]/div[2]/table/tbody/tr/td[6]')
for i in run:
    runs=i.text
    Runs.append(runs)
    
hs=driver.find_elements(By.XPATH,'//*[@id="main-container"]/div[4]/div[1]/div[3]/div[1]/div[1]/div/div[1]/div[2]/table/tbody/tr/td[7]')
for i in hs:
    High=i.text
    High_Score.append(High)
    
avg=driver.find_elements(By.XPATH,'//*[@id="main-container"]/div[4]/div[1]/div[3]/div[1]/div[1]/div/div[1]/div[2]/table/tbody/tr/td[7]')
for i in avg:
    av=i.text
    Average.append(av)

stre=driver.find_elements(By.XPATH,'//*[@id="main-container"]/div[4]/div[1]/div[3]/div[1]/div[1]/div/div[1]/div[2]/table/tbody/tr/td[10]')
for i in stre:
    strr=i.text
    Strike_Rate.append(strr)
    
shun=driver.find_elements(By.XPATH,'//*[@id="main-container"]/div[4]/div[1]/div[3]/div[1]/div[1]/div/div[1]/div[2]/table/tbody/tr/td[11]')
for i in shun:
    hund=i.text
    Hundreds.append(hund)
    
sfif=driver.find_elements(By.XPATH,'//*[@id="main-container"]/div[4]/div[1]/div[3]/div[1]/div[1]/div/div[1]/div[2]/table/tbody/tr/td[12]')
for i in sfif:
    fifty=i.text
    Fifties.append(fifty)
    
four=driver.find_elements(By.XPATH,'//*[@id="main-container"]/div[4]/div[1]/div[3]/div[1]/div[1]/div/div[1]/div[2]/table/tbody/tr/td[14]')
for i in four:
    forr=i.text
    Total_Fours.append(forr)
    
six=driver.find_elements(By.XPATH,'//*[@id="main-container"]/div[4]/div[1]/div[3]/div[1]/div[1]/div/div[1]/div[2]/table/tbody/tr/td[15]')
for i in six:
    sixx=i.text
    Total_Sixes.append(sixx)
```

```python
DataFrame=pd.DataFrame({'Player_name':Player_name,'Span':Span,'Matches':matches,'Innings':Innings,'Not Out':NotOut
                       ,'Runs':Runs,'High_Score':High_Score,'Average':Average,'Strike Rate':Strike_Rate,'100s':Hundreds,
                       '50s':Fifties,'4s':Total_Fours,'6s':Total_Sixes})
```
